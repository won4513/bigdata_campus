```python
gulist = boundary['SIGUNGU_NM'].unique()
```

```python
gulist
```

```python
j = 1
plt.figure(figsize = (30,30))
for i in gulist:
    gu_boundary = boundary[boundary['SIGUNGU_NM'] == i]
    gu_geo = geo19[geo19['ADDR_KOR'].notnull() & geo19['ADDR_KOR'].str.contains(i)]
    ax = plt.subplot(5,5,j)
    gu_boundary.plot(figsize=(5,5), color='whitesmoke', edgecolor='black', linewidth=1,ax = ax)
    ax.axis('off')
    gu_geo.plot(ax=ax, color='deepskyblue', markersize=30)
    plt.title("2019 " + i +'   ')
    j = j+1
plt.show()
```
