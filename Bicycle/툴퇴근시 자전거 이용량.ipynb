```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rc('font', family='Malgun Gothic')
sns.set(font="Malgun Gothic",#"NanumGothicCoding", 
        rc={"axes.unicode_minus":False}, # 마이너스 부호 깨짐 현상 해결
        style='darkgrid')
```

```python
df=pd.read_csv("C:/Users/guswn/OneDrive/바탕 화면/bike/통근·통학시+이용하는+교통수단_20231029143343.csv",encoding="utf-8")
```

```python
# 세번째 줄을 HEADER로 설정
df.columns = df.iloc[0]
df = df[1:]
```

```python
df1=df[df["구분별(1)"]=="지역소분류별"]
```

```python
numeric_columns = df1.columns[2]  # 숫자로 변환할 열 선택, 첫 번째 열은 제외
df1[numeric_columns] = df1[numeric_columns].astype(float)
```

```python
df1=df1.drop("구분별(1)",axis=1)
```

```python
df1.sort_values(by="자전거",ascending=False)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='구분별(2)', y='자전거', data=df1, palette='viridis')
plt.xlabel('구')
plt.ylabel('자전거 사용량')
plt.title('서울 구별 출퇴근시 자전거 사용량')
plt.xticks(rotation=45)
plt.show()
```
