```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data2=pd.read_csv("C:/Users/guswn/OneDrive/바탕 화면/bike/서울특별시 공공자전거 이용정보(월별)_22.07_22.12.csv",encoding="cp949")
data1=pd.read_csv("C:/Users/guswn/OneDrive/바탕 화면/bike/서울특별시 공공자전거 이용정보(월별)_23.1-6 (1).csv",encoding="cp949")
```

```python
#열 이름 동일하게 바꿈
data2.rename(columns={'이동거리(M)': '이용거리(M)','대여일자':'대여년월'}, inplace=True)
```

```python
data=pd.concat([data1,data2])
```

```python
data
numeric_columns = data.columns[7:]  # 숫자로 변환할 열 선택, 첫 번째 열은 제외
data[numeric_columns] = data[numeric_columns].apply(pd.to_numeric, errors='coerce')
```

```python

Q1 = data['이용거리(M)'].quantile(0.25)
Q3 = data['이용거리(M)'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
data = data[(data['이용거리(M)'] >= lower_bound) & (data['이용거리(M)'] <= upper_bound)]
```

```python
sns.boxplot(x=data['이용거리(M)'])
plt.show()
```

```python
data=data[data["이용거리(M)"] != 0]
```
